```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
plt.plot(x_axis,y_axis,color="blue", linewidth=2, marker="*", label="Boston")
plt.xlabel("Date")
plt.ylabel("saving ($)")
plt.ylim(0, 45)
plt.title("Saving Acount Results")
plt.grid()
plt.legend()
```

```python
plt.bar(x_axis,y_axis, color="green", label="boston" )
plt.xlabel("month")
plt.ylabel("Fare($)")
plt.ylim(0,50)
plt.title("PyBar chart bar")
plt.legend()
```

```python
fig,ax=plt.subplots()
ax.bar(x_axis,y_axis, color="black", label="New York")
ax.set_xlabel("months")
ax.set_ylabel("Fare")
ax.legend()
ax.invert_xaxis()
```

```python
plt.scatter(y_axis,x_axis,c="red", s=100,label="Chicago")
plt.title("PayBay By Month")
plt.xlabel("Fare($)")
plt.ylabel("Months")
plt.legend()
plt.gca().invert_yaxis()
```

```python
plt.scatter(y_axis,x_axis, s=[i*5 for i in y_axis], c='blue', linewidth=2, alpha=0.2, label="chicago")
plt.xlabel("fare")
plt.ylabel("months")
plt.title("PayBar")
```
